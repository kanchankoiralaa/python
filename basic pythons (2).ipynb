```python
def analyze(arr):
  
  mean=np.mean(arr)
  median=np.median(arr)
  min=np.min(arr)
  max=np.max(arr)
  print("Minimum:{}\nMaximum:{}\nMean:{}\nMedain:{}".format(min,max,mean,median))

analyze(np.array([1,5,2,4,3]))
```

```python
def trigonometric_plot(x, plot_type):
  if(plot_type == "sin"):
    y=np.sin(x)
  elif(plot_type == "cos"):
    y=np.cos(x)
  else:
    y=np.tan(x)
  mp.title(" Graphical Representation of {} function".format(plot_type))
  mp.plot(x,y)
  mp.show()

my_array = np.linspace(-5,5,75)
trigonometric_plot(my_array,"sin")

my_array = np.linspace(-5,5,75)
trigonometric_plot(my_array,"cos")
```
